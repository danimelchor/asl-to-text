```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib widget
```

```python
model_type = "abc"
```

```python
with open(f'../data/raw/{model_type}_points.json', 'r') as f:
    points = json.load(f)

df = pd.DataFrame(points).reset_index(drop=True)
df.head()
```

```python
from mpl_toolkits import mplot3d

def plot_hand(points):
    line_1 = np.array([points[0], points[1], points[2], points[3], points[4]])
    line_2 = np.array([points[0], points[5], points[6], points[7], points[8]])
    line_3 = np.array([points[9], points[10], points[11], points[12]])
    line_4 = np.array([points[13], points[14], points[15], points[16]])
    line_5 = np.array([points[0], points[17], points[18], points[19], points[20]])
    line_6 = np.array([points[5], points[9], points[13], points[17]])
    lines = [line_1, line_2, line_3, line_4, line_5, line_6]

    fig = plt.figure()
    ax = plt.axes(projection='3d')

    # Scatter plot
    ax.scatter3D(points[:,0], points[:,1], points[:,2], c="red", s=20)

    # Draw lines
    for line in lines:
        ax.plot3D(line[:,0], line[:,1], line[:,2], 'gray')

    # Set axis
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()
```

```python
# Get a matrix representing the letter A
a = df[df['label'] == 'a'].iloc[0]['data']
a = np.array(a)[21:]
plot_hand(a)
```
